```python
import matplotlib.pyplot  as plt

plt.scatter(
    diabetes.data[:,2],
    diabetes.target
)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# 예측값(y^)으로 올바른 모델 찾기
# - 훈련데이터에 잘 맞는 W와 b를 찾는 방법(y^ = wX + b)
# (1) 무작위로 w오 b를 정한다.
# (2) X에서 샘플(첫번째 데이터) 하나를 선택하여  Y^을 계산한다.YouTubeVideo
# (3) Y^과 선택한 샘플의 Y를 비교합니다(예측값과 정답 비교하기 : 손실함수)
# (4) Y^이 Y와 가까워지돌고 w, b를 조정한다. (모델 조정)
# (5) 모든 샘플이 처리될때까지 다시 (2)~(5)항목을 반복 합니다.
```

```python
x = diabetes.data[:,2]
y = diabetes.target
                
```

```python
# 초기의 w, b 정하기
w = 1.0
b = 1.0
```

```python
# 첫번때 샘플로 계산하기
y_hat = w*x[0] + b
y_hat
```

```python
# 실제값
y[0]
```

```python
# 실제값과 가까워지기 위해 w,b를 조금씩 변경해서 y_hat이 증가하는지 감소하는지 파악
# w를 0.1만큼 증가하자
w_inc = w+0.1
y_hat_inc = w_inc * x[0] + b
y_hat_inc
```

```python
# w갑을 조정한 수 예측값 증가 정도 확인  (변화율)
w_rate = (y_hat_inc- y_hat) / (w_inc- w)
w_rate
```

```python
# w_rate = ((w_inc * x[0] + b)) - (w*x[0] + b)) / (w+0.1 - w)
# = x[0] * (w-inc -w) / (w-inc -w)  = x[0]
##변화율은 결국 훈련데이터의 샘플인 x[0] 이다.  
```

```python
# 변화율(Gradient: 경사, 기울기)로 가중치 및 절편 업데이트 하기
w_new = w + w_rate
w_new
```

```python
# b를 0.1만큼  증가시킨후 변화율로 절편 업데이트 하기
b_inc = b + 0.1
y_hat_inc = x[0] + w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat )  / (b_inc - b)
b_rate
```

```python
# b_rate = (y_hat_inc - y_hat )  / (b_inc - b)
# = (x[0] + w + b_inc) - (x[0]*w + b) / (b_inc- b)
# = b_inc - b /   b_inc - b  = 1
# bias의 변화율은 1입니다
```

```python
b_new = b + 1
b_new
```

```python
# 오차 역전파(back propagation)로 가중치와 절편을 더 적절하게 업데이트 하기
# : 오차와 변화율을 곱하여 가중치 업데이트 하기
```

```python
# x[0]
err = y[0] - y_hat
w_new = w + w_rate *err  # 변화율에 오차를 곱한다.
b_new = b + 1*err
print(w_new, b_new)
```

```python
# x[1]
y_hat = x[1]*w_new + b_new
err = y[1] - y_hat

w_rate = x[1]
w_new = w_new + w_rate*err
b_new = b_new + 1*err
print(w_new, b_new)
```

```python
# x[2]
y_hat = x[2]*w_new + b_new
err = y[2] - y_hat

w_rate = x[2]
w_new = w_new + w_rate*err
b_new = b_new + 1*err
print(w_new, b_new)
```

```python
# 전체 샘플을 반복하기
w = 1
b = 1

for x_i , y_i in zip(x,y):
    y_hat = x_i*w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate*err
    b = b+ 1*err

print(w,b)
```

```python
# 산점도에 직선을 넣어 잘 되었는지 확인하기

plt.scatter(
    x,
    y
)


pt1 = (-0.1, -0.1*w + b)
pt2 = (0.15, 0.15*w + b)

plt.plot(
    [pt1[0], pt2[0]],
    [pt1[1], pt2[1]],
    color = 'red'
)
plt.show()
```

```python
# Epoch로 반복하기
: 경사하강법에서는 보통 주어진 훈련데이터로 학습을 여러번 반복합니다.
```

```python
# epoch를 100으로 하자
w = 1
b = 1
for _ in range(100):
    for x_i , y_i in zip(x,y):
        y_hat = x_i*w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate*err
        b = b + 1*err

print(w,b)
```

```python
# 산점도에 직선을 넣어 잘 되었는지 확인하기

plt.scatter(
    x,
    y
)

pt1 = (-0.1, -0.1*w + b)
pt2 = (0.15, 0.15*w + b)

plt.plot(
    [pt1[0], pt2[0]],
    [pt1[1], pt2[1]],
    color = 'red'
)
plt.show()
```
