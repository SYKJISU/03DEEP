```python
# 이미지 만들기
from PIL import Image, ImageDraw
```

```python
image = Image.new('RGB', (169,158), (255,255,255))
draw = ImageDraw.Draw(image)

draw.ellipse(
    (42,42,120,120),
    fill = (0,0,0)
)

image.save("../Data/circle.png")
image
```

```python
# 2개의 이미지 더하기(흑백)
src1 = cv2.imread("../Data/cat.png", cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread("../Data/circle.png", cv2.IMREAD_GRAYSCALE)
```

```python
dst = cv2.add(src1, src2)
fig, axs = plt.subplots(1, 3, figsize=(10 , 5))
axs[0].imshow(cv2.cvtColor(src1, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(src2, cv2.COLOR_BGR2RGB))
axs[2].imshow(cv2.cvtColor(dst , cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 차이
dst = cv2.absdiff(src1, src2)
fig, axs = plt.subplots(1, 3, figsize=(10 , 5))
axs[0].imshow(cv2.cvtColor(src1, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(src2, cv2.COLOR_BGR2RGB))
axs[2].imshow(cv2.cvtColor(dst , cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# CCTV 영상 차이점 비교
# 2개의 이미지 더하기(흑백)
cctv1 = cv2.imread("../Data/cctv01.png", cv2.IMREAD_GRAYSCALE)
cctv2 = cv2.imread("../Data/cctv02.png", cv2.IMREAD_GRAYSCALE)
```

```python
# 이미지 만들기
from PIL import Image, ImageDraw

# CCTV 영상 차이점 비교
# 2개의 이미지 더하기(흑백)
cctv1 = cv2.imread("../Data/cctv01.png", cv2.IMREAD_GRAYSCALE)
cctv2 = cv2.imread("../Data/cctv02.png", cv2.IMREAD_GRAYSCALE)
# 이미지 차이
dst = cv2.absdiff(cctv1, cctv2)
fig, axs = plt.subplots(1, 3, figsize=(10 , 5))
axs[0].imshow(cv2.cvtColor(cctv1, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(cctv2, cv2.COLOR_BGR2RGB))
axs[2].imshow(cv2.cvtColor(dst , cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 
dst = cv2.add(cctv1, cctv2)
fig, axs = plt.subplots(1, 3, figsize=(10 , 5))
axs[0].imshow(cv2.cvtColor(cctv1, cv2.COLOR_BGR2RGB))
axs[1].imshow(cv2.cvtColor(cctv2, cv2.COLOR_BGR2RGB))
axs[2].imshow(cv2.cvtColor(dst , cv2.COLOR_BGR2RGB))
plt.show()
```
