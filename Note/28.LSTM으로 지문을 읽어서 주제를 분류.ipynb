```python
# paragraph에 사용된 중복없는 전체 단어 갯수를 파악
results = set()
df['paragraph'].str.lower().str.split().apply(results.update)
vocab_size = len(results)
vocab_size
```

```python
results
```

```python
# 데이타에서 가장 긴 문장의 단어 갯수를 확인
max_length = 0 
for row in df['paragraph']:
    if len(row.split(" ")) > max_length:
        max_length = len(row.split(" "))

print(max_length)
```

```python
# 단어를 숫자로 인코딩
paragraphs = df['paragraph'].to_list()
paragraphs[0]
```

```python
paragraphs = df['paragraph'].to_list()
paragraphs
```

```python
from tensorflow import keras
keras.utils.set_random_seed(1)
```

```python
for paragraph in paragraphs:
    print(len(paragraph))
```

```python
encoded_paragraphs = [
    keras.preprocessing.text.one_hot(paragraph , vocab_size) 
    for paragraph in paragraphs
]
```

```python
print(encoded_paragraphs)
```

```python
# 문장마다 단어 갯수가 다르므로 , 시퀀스패딩을 넣어서 문자의 길이를 동일하게 맞춘다

padded_paragraphs_encoding = keras.preprocessing.sequence.pad_sequences(encoded_paragraphs, maxlen=max_length, padding='post')
padded_paragraphs_encoding
```

```python
# 분류 항목(food, sports)를 수치로 변경하기
categories = df['category'].to_list()
categories
```

```python
def catetory_encode(category):
    if category == 'food':
        return [1,0]
    else : 
        return [0,1]
```

```python
encoded_category = [catetory_encode(category) for category in categories ]
```

```python
# Feature 확인
print(encoded_paragraphs[0])
print(encoded_paragraphs[19])
```

```python
# Feature 확인
print(padded_paragraphs_encoding[0])
print(padded_paragraphs_encoding[19])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Dense, SimpleRNN , Embedding , LSTM
import tensorflow as tf
from tensorflow import keras
import numpy as np

model = Sequential()

# 문맥생성 단계
model.add(Embedding(vocab_size,5))
model.add(LSTM(64))

# 분류 단계
model.add(Dense(32, activation='relu'))  
model.add(Dense(2, activation='softmax'))  

model.build(input_shape=(None,max_length))

model.summary()

model.compile(optimizer= 'adam' , loss='categorical_crossentropy', metrics=['accuracy'])  

train_X = np.array(padded_paragraphs_encoding)
train_y = np.array(encoded_category)

history = model.fit(
    train_X, train_y,         
    epochs=30, batch_size = 50
)
```

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import one_hot

# 예시 데이터 불러오기
df = pd.read_csv("../Data/lstm.csv")

# 소문자 변환 + 단어 집합 만들기
results = set()
df['paragraph'].str.lower().str.split().apply(results.update)

# 어휘 개수
vocab_size = len(results)

# 단어 집합 기반 인덱스 매핑 (정렬 후 번호 부여)
word_index = {word: i+1 for i, word in enumerate(sorted(results))}

print(f"word_index : {word_index}")
# 비교할 단어
word = "dishplace"

# one_hot 해시 기반 매핑
one_hot_id = one_hot(word, vocab_size)[0]

# results 기반 실제 사전 순 인덱스
index_based_id = word_index[word]

print(f"총 어휘 개수: {vocab_size}")
print(f"단어 '{word}' 의 one_hot 매핑 값: {one_hot_id}")
print(f"단어 '{word}' 의 results 기반 인덱스 값: {index_based_id}")
```
