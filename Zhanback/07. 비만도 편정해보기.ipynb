#### Data 정규화 하기

```python
# weight, height
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
csv.head()
```

```python
# label 
bclass = {
    'thin':[1,0,0],
    'normal':[0,1,0],
    'fat':[0,0,1]
}

y = []

for i, v in enumerate(csv['label']):
    y.append(bclass[v])

y
```

----
##### 훈련용과 테스트 데이터 나누기

```python
# Train
X_train = csv[['height','weight']][:15000]
y_train = y[:15000]
```

```python
# Test
X_test = csv[['height','weight']][15000:]
y_test = y[15000:]
```

----
#### Deep Learning Model
- 입력층
- 은닉층 : 512개의 노드 , activation='relu
- 출력층 

##### 모델 만들기

```python
from tensorflow import keras
from tensorflow.keras.layers import Input
```

```python
model = keras.Sequential()
# 입력층
model.add(
    Input(shape=(2,))
)

# 은닉층
model.add(
    keras.layers.Dense(512, activation='relu')
)

# 출력층
model.add(
    keras.layers.Dense(3, activation='softmax')
)
```

#### 손실함수
- optimizer = 'rmsprop'

```python
model.compile(
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        optimizer='rmsprop'
)
```

#### 데이터 훈련하기
- epochs = 50

```python
import numpy as np

model.fit(
    np.array(X_train),
    np.array(y_train),
    epochs = 49
)
```

#### 테스트 데이터로 평가하기

```python
score = model.evaluate(
            np.array(X_test),
            np.array(y_test)
)
```

```python
print('accuracy =', score[1])
print('loss=', score[0])
```
