### Image 처리

```python
import numpy as np
from PIL import Image
```

```python
# 첫번째 이미지 불러오기
file00 = "../Data/Aiden/image_0000.jpg"
img00 = Image.open(file00)
img00
```

```python
# 두번째 이미지 불러오기
file01 = "../Data/Aiden/image_0001.jpg"
img01 = Image.open(file01)
img01
```

```python
# file00과 file01의 해상도
print(np.array(img00).shape)
print(np.array(img01).shape)
```

```python
# img00을 Gray로 변환후 해상도 확인
img00 = img00.convert('L')
imgArray00 = np.array(img00)
imgArray00.shape
```

```python
# img01을 Gray로 변환후 해상도 확인
img01 = img01.convert('L')
imgArray01 = np.array(img01)
imgArray01.shape
```

```python
# 400 X 300 크기의 Array를 작성해서 기존 이미지 붙히기 

# Empty Array만들기
imgDummy = np.zeros(400*300).reshape(400, 300)

rowNum = (400 - imgArray00.shape[0]) / 2
colNum = (300 - imgArray00.shape[1]) / 2

# Empty List에 Image Array 붙이기
k = 0
for i in list(range(int(rowNum), int(rowNum)+imgArray00.shape[0])):
    l = 0
    for j in list(range(int(colNum), int(colNum)+imgArray00.shape[1])):
        imgDummy[i,j] = imgArray00[k,l]
        l+=1
    k+=1
```

```python
import matplotlib.pyplot as plt

plt.imshow(imgDummy, cmap='gray')
plt.show()
```
